```python
num_fea=[fea for fea in df.columns if df[fea].dtype !="str"]
cat_fea=[fea for fea in df.columns if df[fea].dtype =="str"]
print("we have {} numerical features:{}".format(len(num_fea),num_fea))
print("we have {} categorical features:{}".format(len(cat_fea),cat_fea))
```

```python
df["total_score"]=df["reading_score"]+df["math_score"]+df["writing_score"]
df["average"]=df["total_score"]/3
df.head()
```

```python
reading_full=df[df["reading_score"]==100]["average"].count()
writing_full=df[df["writing_score"]==100]["average"].count()
math_full=df[df["math_score"]==100]["average"].count()
print(f"number of student with full marks in maths:{math_full}")
print(f"number of student with full marks in reading:{reading_full}")
print(f"number of student with full marks in writing:{writing_full}")

```

```python
reading_less=df[df["reading_score"]<=20]["average"].count()
writing_less=df[df["writing_score"]<=20]["average"].count()
math_less=df[df["math_score"]<=20]["average"].count()
print(f"number of student with less marks in maths:{math_less}")
print(f"number of student with less marks in reading:{reading_less}")
print(f"number of student with less marks in writing:{writing_less}")
```

```python
gender_group = df.groupby('gender').mean(numeric_only=True)
gender_group
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average','Math Average']


female_scores = [gender_group.loc['female', 'average'],
                 gender_group.loc['female', 'math_score']]

male_scores   = [gender_group.loc['male', 'average'],
                 gender_group.loc['male', 'math_score']]


X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```
